```python
# 加载数据
with open('structqa_text.json', 'r', encoding='utf-8') as file:  
    json_data = json.load(file)
```

```python
def create_qcae_object(question,choose_list,answer,explanation):
    # 创建一个空的JSON对象
    json_object = {}
    json_object["question"]=question
    # 获取字母表的前N个字母，N为choose_list的长度
    letters = string.ascii_uppercase[:len(choose_list)]

    # 遍历choose_list，并将每个元素分配给对应的字母属性
    for letter, choice in zip(letters, choose_list):
        json_object[letter] = choice

    json_object["answer"] = answer
    json_object["explanation"]=explanation

    return json_object

def split_choose_text(choose_text):
    # 更新正则表达式以直接匹配任意文本，包括换行符
    patterns = [
        r"(?<=A、)(.*?)(?=B、|$)",
        r"(?<=B、)(.*?)(?=C、|$)",
        r"(?<=C、)(.*?)(?=D、|$)",
        r"(?<=D、)(.*)"
    ]

    # 搜索匹配项并编译结果
    matches = []
    for pattern in patterns:
        match = re.search(pattern, choose_text, re.DOTALL)  # 使用re.DOTALL使.匹配包括换行符在内的任意字符
        if match:
            # 应用strip()和replace()到匹配的字符串，而不是re.Match对象
            match_text = match.group(0).replace("\\n", "").replace("\n","")
            # 删除字符串中的 "![任意字符](任意字符)" 部分
            match_text = re.sub(r"!\[.*?\]\(.*?\)", "", match_text).strip()
            matches.append(match_text)
    return matches

# def merge_explanation(explanation):
    



true_false_json_list=[]

# 提取判断题
for item in json_data:
    question = item.get("question", "")
    choose_text=item.get("choose","")
    choose_list=split_choose_text(choose_text)
    answer=item.get("answer","")
    explanation=item.get("explanation","")
    if len(choose_list)==2:
        true_false_json_list.append(item)
        json_object=create_qcae_object(question,choose_list,answer,explanation)
        print(json_object)


```
